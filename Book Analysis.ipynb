```python
import pandas as pd

# Specify the filename of the uploaded dataset (e.g., 'books.csv')
uploaded_file = 'books.csv'  # Update with the actual filename

# Load the dataset using the filename of the uploaded file
df = pd.read_csv(uploaded_file)

# Display the first few rows of the dataset to verify
df.head()
```

```python
print(df.columns) #display the columns of the dataset
```

```python
harry_potter_books = df[df['title'].str.contains('Harry Potter', case=False, na=False)]
print(harry_potter_books)
```

```python
harry_potter_df = df[df['title'].str.contains('Harry Potter', case=False)]
most_selling_books = harry_potter_df.sort_values(by='ratings_count', ascending=False)
print("Most Selling Books in the Harry Potter Series:")
print(most_selling_books[['title', 'ratings_count']])
```

```python
average_rating = harry_potter_df['average_rating'].mean()
print("Average Rating of Harry Potter Books:", average_rating)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have the dataset file named 'books_dataset.csv' in your current directory
books_df = pd.read_csv('books.csv')

# Filter out the Harry Potter books
harry_potter_df = books_df[books_df['title'].str.contains('Harry Potter', case=False)]

# Drop duplicates if any
harry_potter_df.drop_duplicates(inplace=True)

# Reset index
harry_potter_df.reset_index(drop=True, inplace=True)

# Plotting the most selling books within the Harry Potter series
plt.figure(figsize=(10, 6))
sns.barplot(x='title', y='books_count', data=harry_potter_df, estimator=sum, ci=None)
plt.xticks(rotation=45, ha='right')
plt.title('Most Selling Books within the Harry Potter Series')
plt.xlabel('Book Title')
plt.ylabel('Total Books Sold')
plt.tight_layout()
plt.show()

# Calculating the average rating of the Harry Potter books
average_rating = harry_potter_df['average_rating'].mean()

# Plotting the distribution of ratings for the Harry Potter books
plt.figure(figsize=(10, 6))
sns.histplot(data=harry_potter_df, x='average_rating', bins=10, kde=True)
plt.title('Distribution of Ratings for Harry Potter Books')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.axvline(average_rating, color='red', linestyle='--', label='Average Rating')
plt.legend()
plt.show()

# Display the calculated average rating
print("Average Rating of Harry Potter Books:", average_rating)
```
